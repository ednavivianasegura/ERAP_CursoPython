```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Ejercicio 1:
- Gr√°fica una recta de (-7, -7) a (7, 7) de color rosa.

```python
X = np.linspace(-7, 7, 10)
Y = X

plt.plot(X, Y, color = "pink")
plt.show()
```

### Ejercicio 2:
- Obt√©n el siguiente resultado utilizando subplot y la ecuaci√≥n anterior.



```python
X = np.linspace(-7, 7, 10)
Y = X

colores = ["pink", "blue", "red", "magenta"]

for index, color in enumerate(colores, start = 1):

    plt.subplot(2, 2, index)
    plt.plot(X, Y, color = color)

plt.show()
```

### Ejercicio 9:
- Muestra en una grafica de barras el n√∫mero de hombres y mujeres que sobrevivieron.



```python
etiquetas = sorted(titanic["Sex"].unique())
etiquetas
```

```python
fig, ax = plt.subplots()

etiquetas = sorted(titanic["Sex"].unique())

valores1 = titanic[titanic["Survived"] == 1].value_counts("Sex", sort = False).values
valores2 = titanic[titanic["Survived"] == 0].value_counts("Sex", sort = False).values


x = np.arange(len(etiquetas))  
width = 0.4

plt.bar(x - width/2, valores1, width, color = "blue")
plt.bar(x + width/2, valores2, width, color = "red")

ax.set_xticks(x)
ax.set_xticklabels(etiquetas)

plt.show()
```

### Ejercicio 10:
- Muestra en una grafica de barras agrupada la cantidad de hombres y mujeres de cada clase (**`Pclass`**).



```python
fig, ax = plt.subplots(figsize = (10, 6))

etiquetas = sorted(titanic["Sex"].unique())

valores1 = titanic[titanic["Pclass"] == 1]["Sex"].value_counts(sort = False).values
valores2 = titanic[titanic["Pclass"] == 2]["Sex"].value_counts(sort = False).values
valores3 = titanic[titanic["Pclass"] == 3]["Sex"].value_counts(sort = False).values

x = np.arange(len(etiquetas))  

width = 0.4

plt.bar(x + width/1, valores3, 0.2, color = "green", label = "Clase 3")
plt.bar(x - width/3, valores1, 0.2, color = "blue", label = "Clase 1")
plt.bar(x + width/3, valores2, 0.2, color = "red", label = "Clase 2")

plt.legend(loc = "best")

ax.set_xticks(x)
ax.set_xticklabels(etiquetas)

plt.show()
```

### Ejercicio 15:
- Muestra en una grafica de caja la edad de los que sobrevivieron y de los que no sobrevivieron.

**Nota**: Para poder graficarlo hay que reemplazar los NaN por el promedio de edad.



```python
titanic["Age"].replace(np.nan, titanic["Age"].mean(), inplace = True)
```

```python
etiquetas = ["Survived" if i == 1 else "Not Survived" for i in titanic["Survived"].unique()]


data1 = titanic[titanic["Survived"] == 0]["Age"]
data2 = titanic[titanic["Survived"] == 1]["Age"]

data = [data1, data2]

plt.boxplot(x = data, labels = etiquetas)
plt.show()
```
